# Regresión Logística

## Librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.model_selection import  GridSearchCV
from scikitplot.metrics import plot_roc,plot_ks_statistic

import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

## Entrenamiento del modelo

### Lectura de datos

```python
train=pd.read_csv('~/train_discretizado.csv')
train.head()
```

### Mejores variables

```python
best=['woe_disc_tot_cur_bal_10', 'woe_disc_int_rate_10', 'woe_disc_inq_last_6mths_10', 'woe_disc_annual_inc_10', 
      'woe_disc_dti_9', 'woe_disc_mths_between_issued_earliest_cr_10', 'woe_disc_total_rev_hi_lim_10']

target='target'
```

### Entrenamiento

```python
logreg=LogisticRegression()
logreg.fit(train[best],train[target])
```

### Desempeño del modelo en el conjunto de entrenamiento

```python
# porcentaje de aciertos
accuracy_score(train[target],logreg.predict(train[best]))
# resultado alto dado que la muestra esta desbalanceada
```

```python
# area bajo la curva ROC
roc_auc_score(train[target],logreg.predict_proba(train[best])[:,1])
```

```python
plot_roc(train[target],logreg.predict_proba(train[best]),plot_micro=False,plot_macro=False,figsize=(10,4))
```

```python
plot_ks_statistic(train[target],logreg.predict_proba(train[best]),figsize=(10,4))
```

## Hiperparámetros

```python
param_grid={'penalty':['l1', 'l2', None],
            'C':[.001,0.01,0.1,1,10],
            'class_weight':[None,'balanced']
            }
```

```python
gscv=GridSearchCV(logreg,param_grid,scoring='roc_auc',cv=5)
gscv.fit(train[best],train[target])
```

```python
# mejores parametros
gscv.best_params_
```

```python
modelo=gscv.best_estimator_
modelo.fit(train[best],train[target])
```

```python
roc_auc_score(train[target],modelo.predict_proba(train[best])[:,1])
```

```python
plot_roc(train[target],modelo.predict_proba(train[best]),plot_micro=False,plot_macro=False,figsize=(10,4))
```

```python
plot_ks_statistic(train[target],modelo.predict_proba(train[best]),figsize=(10,4))
```
